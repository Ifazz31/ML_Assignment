Importing Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.datasets import make_classification
from sklearn.tree import DecisionTreeClassifier, export_text
from sklearn import tree
```

Generating Synthetic Dataset

```python
X, y = make_classification(n_samples=100, n_features=2, n_informative=2, n_redundant=0, random_state=42)
```

Training Logistic Regression model

```python
logistic_regression_model = LogisticRegression()
logistic_regression_model.fit(X, y)
```

```python
# Plot Logistic Regression Decision Boundary
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)

# Using 'viridis' colormap
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.RdYlBu)
plt.title('Logistic Regression Decision Boundary')

xx, yy = np.meshgrid(np.linspace(X[:, 0].min(), X[:, 0].max(), 100),
                     np.linspace(X[:, 1].min(), X[:, 1].max(), 100))
Z = logistic_regression_model.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

# Using 'plasma' colormap for contourf
plt.contourf(xx, yy, Z, alpha=0.3, cmap=plt.cm.RdYlBu)
plt.show()
```

Decision Tree Model

```python
decision_tree_model = DecisionTreeClassifier()
decision_tree_model.fit(X, y)

# Plot Decision Tree
plt.figure(figsize=(16, 4))

# Scatter plot using 'PuBuGn' colormap
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.PuBuGn)
plt.title('Custom Decision Tree Decision Boundary - PuBuGn Colormap')

xx, yy = np.meshgrid(np.linspace(X[:, 0].min(), X[:, 0].max(), 100),
                     np.linspace(X[:, 1].min(), X[:, 1].max(), 100))
Z = decision_tree_model.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)


plt.contourf(xx, yy, Z, alpha=0.3, cmap=plt.cm.PuBuGn)


tree.plot_tree(decision_tree_model, filled=True, feature_names=['X1', 'X2'], class_names=['Negative', 'Positive'])
plt.title('Decision Tree')
plt.show()
```
